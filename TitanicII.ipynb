# 1. Importar librerias

```python
import os
import pandas as pd
from pandasai import SmartDatalake
from pandasai.responses.streamlit_response import StreamlitResponse
import warnings
```

```python
df = pd.read_csv("titanic.csv", encoding="utf8")

#os.environ["PANDASAI_API_KEY"] = ""

agent = SmartDatalake(
    [df],
    config={"verbose": True, "response_parser": StreamlitResponse},
)
```

```python
agent.chat("Grafico de barras de la cantidad de pasajeros mujeres y hombres")
```

```python
agent.chat("Que cantidad de pasajeros de primera clase habia en el titanic?")
```
