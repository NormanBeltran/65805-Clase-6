```python
import openai
import matplotlib as plt
from io import StringIO
import csv
```

```python
#openai.api_key =""
```

```python
# Datos de construcciones y características
data_str = """Nombre,Altura,Material
Torre,150,Metal
Edificio,80,Cemento
Rascacielos,200,Vidrio
Mansión,30,Vidrio
Shopping,200,Acero
Departamentos,90,Cemento
Casa,80,Metal
"""

# Crear un objeto StringIO para simular un archivo CSV
csv_file = StringIO(data_str)

# Leer el CSV y almacenar los datos en una lista de diccionarios
reader = csv.DictReader(csv_file)
data = list(reader)

# Mostrar los datos
print(data)
```

```python
prompt = "En base a los datos que se encuentran en el contexto, extrae entidades y relaciones, con el objetivo de crear un grafico con los datos, la salida debe ser similar a {'Nombre': 'Torre A', 'Altura': '150', 'Material': 'Metal'}, {'Nombre': 'Edificio B', 'Altura': '80', 'Material': 'Cemento'} solo tomalo como ejemplo"

prepared_prompt = f"Contexto:{data}\n {prompt}"
print(prepared_prompt)
```

```python
# Make the request
conversation = [
        {"role": "system", "content": "Eres un asistente util, tu objetivo principal es extraer todas las relaciones entre las entidades y sus datos,  retorna un array de objetos"},
        {"role": "user", "content": prepared_prompt},
    ]

# Make the request
response = openai.ChatCompletion.create(
    model= 'gpt-3.5-turbo',
    max_tokens=1500,
    messages=conversation
    )

message = response.choices[0]['message']
print("{}: {}".format(message['role'], message['content']))
```

```python
import networkx as nx

G = nx.MultiDiGraph()

data = eval(message['content'])
print(f"data {data}")
#data = eval(str(data))

for d in data:
    G.add_edge(d['Entidad A'], d['Entidad B'], **d)
```

```python
pos = nx.spring_layout(G, k=0.7)  # layout spring for the nodes position
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=5, edge_color='gray', width=1, alpha=0.8)

# Etiquetas de los bordes
edge_labels = {(n1, n2): data['Relacion'] for n1, n2, data in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=5)
```
